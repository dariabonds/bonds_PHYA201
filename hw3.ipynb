# Radiative Processes: HW3

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# initial values for constants
k = 1                # Coulomb constant [kg m^3 s^-4 A^-2]
q_1 = -1             # charge of electron
q_2 = 1              # charge of ion
m_e = 1              # mass of electron
m_i = 10000          # mass of ion
dt = 0.01            # time step [s]
N = 1000             # number of time steps

# position of ion
ion = [0,0]          # positioned at origin

# Bohr radius
a0 = 5.29*10**(-11) # [m]

# time
t_n = np.arange(0, 10.0, 0.01)
```

```python
# need to put the x and y axes in terms of bohr radius
# need to adjust the starting velocities (reference instructions)
# need to adjust the number of time steps (reference instructions)
# a fourier transform of the acceleration to determine the power spectrum of the resulting radiation as a function of frequency
```

### value set 1

```python
# initialize arrays
## for position, velocity, and acceleration
r = np.zeros([N, 2])
v = np.zeros([N, 2])
a = np.zeros([N, 2])

# initialize values
## for initial position and velocity
r_0 = [-10, 1]
v_0 = [1, 0]

r[0] = r_0
v[0] = v_0
```

```python
# iterate calculations
## for the change in time for the number of timesteps
for i in range(N):
    # calculate acceleration
    a[i] = (k * q_1 * q_2) / (m_e * np.linalg.norm(r[i])**3) * r[i]
    
    # break at the final position
    if i == N - 1:
        break
    
    # calculate velocity
    v[i + 1] = v[i] + a[i] * dt
    
    # calculate position
    r[i + 1] = r[i] + v[i] * dt + 0.5 * a[i] * dt**2
    
```

```python
# extract x and y components
## for position, velocity, and acceleration
r_x = r[:, 0]
r_y = r[:, 1]
v_x = v[:, 0]
v_y = v[:, 1]
a_x = a[:, 0]
a_y = a[:, 1]
```

```python
# plotting the position
plt.scatter(x=0, y=0, color='yellow')
plt.scatter(r_x, r_y)
plt.title('position of electron 1')
plt.show
```

```python
# plotting the velocity
plt.scatter(t_n, v_x, color='red')
plt.scatter(t_n, v_y, color='green')
plt.title('velocity components vs time 1')
plt.show
```

```python
# plotting the acceleration
plt.scatter(t_n, a_x, color='orange')
plt.scatter(t_n, a_y, color='purple')
plt.title('acceleration vs time 1')
plt.show
```

```python
# calculate the magnitude of the acceleration
a_mag = np.sqrt(a_x * a_x + a_y * a_y)
```

```python
# Fourier transform of the magnitude of acceleration
# shift the plot so that the peak is aligned
fft_data = np.fft.fftshift(np.fft.fft(a_mag))
```

```python
# define the dimensions of the frequency
start = len(fft_data)
stop = 1 / dt
# shift the frequency do that the peak is aligned
freq = np.fft.fftshift(np.fft.fftfreq(start, stop))
# plot the frequency vs the absolute value of the Fourier transform squared
plt.plot(freq, np.abs(fft_data) * np.abs(fft_data))
plt.title('frequency vs Fourier transform 1')
```

```python
# zoom in on the maximum relevant angular velocity
plt.plot(freq, np.abs(fft_data) * np.abs(fft_data))
plt.title('zoom in 1')
plt.xlim(-0.0001, 0.0003)
```
